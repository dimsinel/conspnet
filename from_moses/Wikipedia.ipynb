```python
docs1_d={}
emptydocs = {}
for i,q in enumerate(nodes):
    qq=revco(q)
    if qq in changes:
        print(f'# {i}: {qq} ---> {changes[qq]}')
        qq = changes[qq]
    #print(i, qq)
    page_py = wiki_wiki.page(qq)
    
    if page_py.exists(): 
        docs1_d[qq]=page_py.summary
    else:
        emptydocs[qq] = wikipedia.search(qq)
        #print(i, qq)
        docs1_d[qq]='  '
    if i%50==0:
        print('> ',i,qq)
        
print(len(docs1_d))    
```
